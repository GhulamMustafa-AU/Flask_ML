```python
import pandas as pd
import numpy as numpy
import pickle
from sklearn.linear_model import LogisticRegression
```

```python
url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
df = pd.read_csv(url)
df
```

```python
{"SibSp":df["SibSp"].isnull().sum(), "Survived":df["Survived"].isnull().sum(), "Pclass":df["Pclass"].isnull().sum(), "Name":df["Name"].isnull().sum(), "Sex":df["Sex"].isnull().sum(), "Age":df["Age"].isnull().sum(), "Parch":df["Parch"].isnull().sum(), "Ticket":df["Ticket"].isnull().sum(), "Fare":df["Fare"].isnull().sum(), "Cabin":df["Cabin"].isnull().sum(), "Embarked":df["Embarked"].isnull().sum()}
```

```python
include = ['Age', 'Sex', 'Embarked', 'Survived'] 
df = df[include]
df
```

```python
catagorical = []
for col, col_type in df.dtypes.items():
    if col_type == "object":
        catagorical.append(col)
    else:
        df[col].fillna(0, inplace=True)
```

```python
df = pd.get_dummies(df, columns=catagorical,dummy_na=True)
```

```python
df
```

```python
Y = df["Survived"]
X = df[df.columns.difference(["Survived"])]
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X, Y)
```

```python
pickle.dump(lr, open("./FlaskAPI/model.pkl", "wb"))
```
